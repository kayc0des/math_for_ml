```python
# Solve for Critical Point
# the critical point is the point where in dy/dx = 0

critical_points = solve(dy_dx(f(x)), x)
print("Critical Points:", critical_points)
```

# Task 4: Create separate array of Minimas and maximas

```python
second_derivative = diff(dy_dx(f(x)), x)

arr_minimas = []
arr_maximas = []

for point in critical_points:
    concavity = second_derivative.subs(x, point)
    if concavity > 0:
        arr_minimas.append(point)
    elif concavity < 0:
        arr_maximas.append(point)

print(f'the array of minimas {arr_minimas} and the array of maximas is {arr_maximas}')
```

#Finally, What is the global Minima and the Global maxima _ Plot this so that I can see

```python
def quadratic_function(x):
    return -x**2 + 4*x + 5

# Generate x values
x_values = np.linspace(-1, 5, 100)

# Calculate corresponding y values using the quadratic function
y_values = quadratic_function(x_values)

# Find the x-coordinate of the vertex (global maximum)
x_vertex = -symbols('b') / (2 * symbols('a'))

# Substitute the values of coefficients into the vertex formula
x_vertex_value = x_vertex.subs({symbols('a'): -1, symbols('b'): 4})
y_vertex_value = quadratic_function(x_vertex_value)

# Plot the quadratic function
plt.plot(x_values, y_values, label='f(x) = -x^2 + 4x + 5')
plt.scatter(x_vertex_value, y_vertex_value, color='red', label='Global Maximum')

# Highlight the vertex (global maximum)
plt.annotate(f'({x_vertex_value},{y_vertex_value})', xy=(x_vertex_value, y_vertex_value),
             xytext=(x_vertex_value + 0.5, y_vertex_value - 5),
             arrowprops=dict(facecolor='black', shrink=0.05))

# plot labels
plt.title('Quadratic Function with Global Maximum')
plt.xlabel('x')
plt.ylabel('f(x)')

# Display grid and legend
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()

plt.show()
```
