```python
ids = data['id']
co2 = data['co2']

plt.hist(co2, bins=10, edgecolor='white')

plt.title('CO2 Distribution')
plt.xlabel('CO2 Value')
plt.ylabel('Number of Cars')

median_value = np.median(co2)
plt.axvline(median_value, color='red', linestyle='dashed', linewidth=2, label='Median')

plt.legend()
plt.tight_layout()

plt.show()
```

From Monday's class, the histogram above is a symmetrical histogram. This symmetry could defined as one where data values are distributed equally on both sides of the central point. The median is the median value of the co2 column.

A greater percentage of cars have co2 values between 300 - 420. As we move further left (towards 0) and further right (towards 700) the number of cars decrease.

Does perfect symmetry exist?

```python
cylinders = data['cylinders']

plt.hist(cylinders, bins=10, edgecolor='white', log=True) #log scale to allow for more visibility with marginal values

plt.title('Cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel('Number of Cars')

plt.tight_layout()

plt.show()
```

```python
pv2 = data['pv2']

plt.hist(pv2, bins=10, edgecolor='white')

plt.title('pv2')
plt.xlabel('pv2 value')
plt.ylabel('Number of Cars')

plt.tight_layout()

plt.show()
```

## **Interpretation**

Th seaborn could be interpreted as:

- Year Column: This shows a light brown color with a correlation around 0.25 with the ghgscore. This suggests a relatively weak positive correlation, indicating that, over time, there may be a slight increase in greenhouse gas scores. However, the correlation is not strong, so the impact of the manufacturing year on the 'ghgscore' may be limited.


- Cylinders and Display Columns: Both columns show a strong negative correlation with the ghgscore. This implies that vehicles with fewer cylinders and a particular display feature tend to have lower greenhouse gas scores. Consideration should be given to the potential environmental benefits associated with these characteristics.


- PV2 and PV4 Columns: The pv2 column shows a light brown color suggesting a weak positive correlation with ghgscore. On the other hand, pv4 has a correlation of near 0, indicating no apparent correlation with ghgscore. These variables may have limited impact on greenhouse gas scores.


- City, UCity, Highway, UHighway, Comb, FEScore: These columns all show a green color with a correlation of 1 or near 1. This strong positive correlation indicates that higher values in these variables are associated with higher ghgscores.


- CO2 Column: This shows a negative correlation indicated by a brown color. This implies that higher CO2 emissions are associated with lower ghgscores.
